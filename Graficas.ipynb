```python
import numpy as np
np.random.seed(1)
x = data_n[:,:3]
d = data_n[:,None,3]
nd = x.shape[0]
ne = x.shape[1]
w = np.random.rand(ne,1)
print(x)
print(d)
print(w)
y = np.zeros((nd,1))
nit = 100
ecm = np.zeros((nit,1))
```

```python
x.shape[0]
```

```python
for k in range(nd):
  a = x[k,None,:] @ w
  y[k,:]=a
```

```python
fig =px.line(y=[y[:,0],d[:,0]])
fig.show()
```

```python
print(f"Los w iniciales son: {w}")
for i in range(nit):
  for k in range(nd):
    a = x[k,None,:] @ w
    y[k,:]=a
    error = y[k,None,:] - d[k,None,:]
    w = w-0.6*error*1*x[k,None,:].T
    print(f"Los w de la repetición {i+1} en el lado {k+1} son: {w}")
    ecm[i,:] = ecm[i,None,:] + 1/(2*nd)*error**2
```

```python
fig =px.line(y=[y[:,0],d[:,0]])
fig.show()
```

```python
x=datos_n.to_numpy()
d=datos[['target']].to_numpy()
nd = x.shape[0]
ne = x.shape[1]
w = np.random.rand(ne,1)
y = np.zeros((nd,1))
nit = 100
ecm = np.zeros((nit,1))
```

```python

for i in range(nit):
  for k in range(nd):
    a = x[k,None,:] @ w
    print(a)
    y[k,:]=1/(1+np.exp(-a))
    error = y[k,None,:] - d[k,None,:]
    w = w-0.3*error*y[k,None,:]*(1-y[k,None,:])*x[k,None,:].T

    ecm[i,:] = ecm[i,None,:] + 1/(2*nd)*error**2
```

```python
fig = go.Figure(
    data = [
            go.Scatter(y=ecm[:,0])
            ]
    )
fig.show()
```
