```python
datos = pd.DataFrame({'Matricula':matricula,
                      'Matematicas':matematicas})
datos.head()
```

```python
#Explorar los datos 
datos.Matricula.value_counts().sort_index()
```

```python
#Gráfico de violin 
fig,ax = plt.subplots(1,1,figsize=(16,9))
sb.violinplot(x='Matricula', y ='Matematicas', data=datos,ax=ax)
ax.set_title('Distribución de notas por clase')
```

```python
##preparación de los datos 
x=datos[['Matematicas']]
y = datos[['Matricula']]

xEntrenamiento, xPrueba, yEntrenamiento, yPrueba = train_test_split(x,y,train_size=0.7, random_state=123)
```

```python
datosEntrenamiento = pd.concat([xEntrenamiento,yEntrenamiento],axis=1)
```

```python
#Gráfico de violin 
fig,ax = plt.subplots(1,1,figsize=(16,9))
sb.violinplot(x='Matricula', y ='Matematicas', data=datosEntrenamiento,ax=ax)
ax.set_title('Distribución de notas por clase')
```

```python
datosPrueba = pd.concat([xPrueba,yPrueba],axis=1)
```

```python
#Gráfico de violin 
fig,ax = plt.subplots(1,1,figsize=(16,9))
sb.violinplot(x='Matricula', y ='Matematicas', data=datosPrueba,ax=ax)
ax.set_title('Distribución de notas por clase')
```

```python
##Modelado de datos 
modelo = LogisticRegression()
modelo.fit(xEntrenamiento,yEntrenamiento)
```

```python
print('Intercepto:', modelo.intercept_)
print('Coeficiente:', list(zip(x.columns,modelo.coef_.flatten())))
print('Desempeño en el entrenamiento:', modelo.score(xEntrenamiento,yEntrenamiento))
```
